# **Simple Moving Average (SMA)**

    1. statistical measure in time series analysis, particularly in the context of stock price data
    2. It is used to smooth out short-term fluctuations and highlight longer-term trends or cycles
    3. we use rolling windows to calculate SMA is because it allows us to compute the average of the
                    data points within a specific window of time as it moves through the dataset
    4.Noise refers to random variations or fluctuations in the data that cannot be attributed to any underlying patterns or trends

# **Rolling**

## **Checking Data is Stataionary or Not**

```python
def adf_test(series):
    result=adfuller(series)
    print("ADF Statistics: {}".format(result[0]))
    print("P-Value: {}".format(result[1]))
    if result[1]<=0.05:
        print("Strong Evidence against null hypothesis,Reject the Null hypothesis,Data has not root and is Stationary")
    else:
        print("Weak Evidence against null hypothesis,Time series has root unit,indicating it is non stationary")
```

```python
adf_test(airline['Thousands of Passengers'])
```

## **Making the data Stationary For Predictions**

```python
# use some techincal Difference
#shift(1)-->because only one position will be shifted down
airline['Passenger First Difference'] = airline['Thousands of Passengers']-airline['Thousands of Passengers'].shift(1)
```

```python
airline.head()
```

```python
adf_test(airline['Passenger First Difference'].dropna())
```

### **2nd Difference for Stationary Data**

```python
airline['Passenger 2nd Difference'] = airline['Passenger First Difference']-airline['Passenger First Difference'].shift(1)
```

```python
airline.head()
```

```python
adf_test(airline['Passenger 2nd Difference'].dropna())
```

### **USe entire year data to check Stationary**

```python
airline['Passenger 12 Difference'] = airline['Thousands of Passengers']-airline['Thousands of Passengers'].shift(12)
```

## **Train test split**

```python
train_dataset_end =datetime(1955,12,1) 
test_dataset_end = datetime(1960,12,1)
```

```python
train_data = airline[:train_dataset_end]
test_data = airline[train_dataset_end+timedelta(days=1):test_dataset_end]
```

```python
pred_start_date=test_data.index[0]
pred_end_date=test_data.index[-1]
```

```python
test_data
```

## **Create an ARIMA Model**

```python
train_data
```

```python
model_arima = ARIMA(train_data['Thousands of Passengers'],order=(4,2,3))
```

### **How to specify the order**

    1.Specifying the order of an ARIMA model involves choosing three parameters: ,q,d.
        1.p - Autoregressive (AR) order
        2.q - Moving Average (MA) order
        3.d - Differencing order

```python
model_arima_fit = model_arima.fit()
```

```python
model_arima_fit.summary()
```

```python
pred_start_date=test_data.index[0]
pred_end_date=test_data.index[-1]
print(pred_start_date)
print(pred_end_date)
```

```python
predictions = model_arima_fit.predict(start=pred_start_date,end=pred_end_date)
residuals = test_data['Thousands of Passengers']-predictions
```

```python
residuals
```

```python
predictions
```

```python
model_arima_fit.resid.plot(kind='kde')
```

```python
test_data['predicted_ARIMA']=predictions
```
